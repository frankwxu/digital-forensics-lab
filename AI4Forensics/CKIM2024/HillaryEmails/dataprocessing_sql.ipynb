```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("archive/database.sqlite")

# Get cursor object
cursor = conn.cursor()

# Query to list all tables
query = "SELECT name FROM sqlite_master WHERE type='table';"

# Execute the query
cursor.execute(query)

# Fetch all tables
tables = cursor.fetchall()

# Print the list of tables
print("Tables in the SQLite database:")
for table in tables:
    print(table[0])

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("archive/database.sqlite")

# Get cursor object
cursor = conn.cursor()

# Query to fetch schema of the "emails" table
query = "PRAGMA table_info(emails);"

# Execute the query
cursor.execute(query)

# Fetch all records
schema = cursor.fetchall()

# Print the schema
for column in schema:
    print(column)

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("archive/database.sqlite")

# Get cursor object
cursor = conn.cursor()

# Query to fetch rows containing the keyword "Israel" in the "ExtractedBodyText" column
query = "SELECT ExtractedBodyText FROM emails WHERE lower(ExtractedBodyText) LIKE '%israel%';"

# Execute the query
cursor.execute(query)

# Fetch all rows
rows = cursor.fetchall()

# Define the file name
file_name = "email_israel.txt"

# Open the file in write mode
with open(file_name, "w") as file:
    # Write each row to the file
    for row in rows:
        file.write(str(row) + "\n")


# Close the connection
conn.close()
```
